```python
import seaborn as sns
import matplotlib as plt
import numpy as np
import pandas as pd
import json
import csv
import nltk
```

```python
h=pd.read_csv('C:/Users/Administrator/Desktop/feedback/R Analysis Test_excel data_2019_20.csv')
h.head()
```

```python
stopwords = nltk.corpus.stopwords.words('english')
stemmer = nltk.stem.PorterStemmer()
docs = h['How can we imporve your outcomes using Automation?'].str.lower().str.replace('[^a-z ]','')
def clean_sentence(text):
    words=text.split(' ')
    words_clean=[stemmer.stem(word) for word in words if word not in stopwords]
    return ' '.join(words_clean)
docs_clean = docs.apply(clean_sentence)
docs_clean.head()
```

```python
all_text=list(docs_clean)
all_text
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(1,1), stop_words = 'english') # we can define our own parameters
X = cv.fit_transform(all_text)
```

```python
Xc = (X.T * X) # This is the matrix manipulation step
Xc.setdiag(0) # We set the diagonals to be zeroes as it's pointless to be 1
```

```python
import pandas as pd
names = cv.get_feature_names() # This are the entity names (i.e. keywords)
df = pd.DataFrame(data = Xc.toarray(), columns = names, index = names)
df.to_csv('to gephi2.csv', sep = ',')
```
